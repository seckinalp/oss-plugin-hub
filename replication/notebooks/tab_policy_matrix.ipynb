- Data sources:

- Expected output:
  - Table: Condensed marketplace policy matrix (tab:policy_matrix)


```python
from pathlib import Path
import json
import pandas as pd

DATA_DIR = (Path.cwd() / ".." / "data").resolve()
policy_path = DATA_DIR / "policy-matrix.json"

with policy_path.open("r", encoding="utf-8") as handle:
    policy_data = json.load(handle)

rows = policy_data["rows"]

df = pd.DataFrame(rows, columns=[
    "platform",
    "review_type",
    "source_packaging_policy",
    "primary_trust_signal",
])

df = df.rename(columns={
    "platform": "Platform",
    "review_type": "Review Type",
    "source_packaging_policy": "Source / Packaging Policy",
    "primary_trust_signal": "Primary Trust Signal",
})

df
```
